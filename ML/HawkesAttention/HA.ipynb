## Hawkes-inspired Attention: Architecture reproduction

<ins> Source <ins>: 

- https://arxiv.org/pdf/2601.09220

<ins> Motivation <ins>:

- For a multivariate Hawkes, one has $\lambda_t = \int_{(0,t)} \phi(t-s) dN_s$ with $\phi$ a $\mathcal{M}_{n}(\mathbb{R})$-valued function. This reads off as $\lambda^{c_i}_t = \sum_{j=1}^n \int_{(0,t)} \phi_{c_i, c_j}(t-s) dN^{c_j}_s$.

- We assume the following separability assumption that $\phi_{c_i, c_j}(t) := \beta_{c_i, c_j} \Phi_{c_j}(t)$, with the following representation of $\beta = UV^T$ with $U \in \mathcal{M}_{n\times d}(\mathbb{R})$ and $V \in \mathcal{M}_{n \times d}(\mathbb{R})$. As such, when $d << n$, this corresponds to a low-rank $\beta$ assumption.

- In this form, one has $\phi = UV^T \Phi$ with $\Phi$ diagonal and then, $\lambda_t = UV^T \Phi \ast dN$. This expression of the intensity ressembles the attention pattern with query $U$, key-value pair $(\Phi \ast dN)^T V$. 

- The idea here here to use the entries of $V^T \Phi \ast dN$ as being the time-modulated embeddings that will serve us for the representation at time $t_j$.

- The resulting attention layer is built from these time-modulated embeddings with the following shape for queries, values and keys:
    - $Q_{j,k} = W_Q v_{c_j}\Phi_{c_j}(t_j - t_k)$.
    - $K_{k,j} = W_K v_{c_k}\Phi_{c_k}(t_j - t_k)$.
    - $V_{k,j} = W_V v_{c_k}\Phi_{c_k}(t_j - t_k)$.

- Unlike TransformerHawkesProcess, the hidden representation is then built directly by an aggregate of the SoftMax dot-products: $h(t_j) := \sum_{t_k < t_j} SoftMax(\frac{Q_{j,k} K_{k,j}^T}{\sqrt{d}}) V_{k,j}$.

- From this hidden state representation, we get the resulting $\lambda(t) = SoftPlus(\mu + a^T h(t))$.
